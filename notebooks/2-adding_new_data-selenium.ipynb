```python
types_one = []
for l in data_one.URL:
    try:
        driver.get(l)
        txt = driver.find_element_by_class_name('css-pz80c5').text
        split_text = txt.split('\n\n')
        types_one.append(split_text)
    except:
        types_one.append('All')
```

```python
types_one
```

```python
skin_types = pd.DataFrame({'col':types_one})
```
